```python
# Utilizar pacotes de biblioteca utilizado pelo Python
import pandas as pd
```

```python
# Realizar a leitura e Visualização do Banco de Dados (bd)
bd = pd.read_csv("/content/An-lise-de-Dados/ClientesBanco.csv" , encoding="latin1")
display(bd)
```

**AC2 - Tratar a Base de Dados e Ajuste de Visualização de Dados dos pandas;**

```python
# Para fazermos uma análise assertiva sobre o cancelamento do cartão, algumas colunas serão excluidas
# No caso será excluídos as colunas: CLIENTNUM,
bd = bd.drop("CLIENTNUM", axis=1)
display(bd)
```

```python
# Compreender as informações que a tabela possui, faz um mapeamento. Importante ver se tem alguma linha faltante.
display(bd.info())
```

```python
# Percebe-se que possuem 10127 linhas estão preenchidas 100% (10127 clientes), com exceção da linha 7 tem-se 10126 (faltando informações).
# Como esta linha pode não fazer diferença sobre o total a ser analisado, podemos exclui-la através do código abaixo (excluir um cliente)
bd = bd.dropna()
display(bd.info())
```
